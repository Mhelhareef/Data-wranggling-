## creating 'tweet_json.txt' 

```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer

# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
# These are hidden to comply with Twitter's API terms and conditions
consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)

# NOTE TO STUDENT WITH MOBILE VERIFICATION ISSUES:
# df_1 is a DataFrame with the twitter_archive_enhanced.csv file. You may have to
# change line 17 to match the name of your DataFrame with twitter_archive_enhanced.csv
# NOTE TO REVIEWER: this student had mobile verification issues so the following
# Twitter API code was sent to this student from a Udacity instructor
# Tweet IDs for which to gather additional data via Twitter's API
tweet_ids = df_1.tweet_id.values
len(tweet_ids)

# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

## writing ''tweet_json.txt' into pandas dataframe

```python
list1 = []
with open('tweet_json.txt') as file: 
  
    for line in file: 
        data = line.split(', ')
        for item in data:
            if  "id_str" in item:
                tweet_id = item.split(': ',1)[1].replace('"','')
                break
        for item in data:
            if "retweet_count" in item:
                retweet_count = int(item.split(': ',1)[1])
                break
        for item in data:
            if "favorite_count" in item:
                favourites_count = int(item.split(': ',1)[1])
                break
        dict1 = {'tweet_id':tweet_id, 'retweet_count':retweet_count, 'favourites_count':favourites_count}
        list1.append(dict1)
        
df_3 = pd.DataFrame(list1)
df_3.head()
```

## Cleaning Tidiness issues:

### copy df_1 to a new data frame to make changes on.

```python
df1_clean = df_1.copy()
#test
#view the head of the copied df 
df1_clean.head()
```

## drop irrelevant columns
### use drop method to remmve unwanted columns

```python
df1_clean.drop(['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id',
                'retweeted_status_user_id', 'retweeted_status_timestamp', 'rating_denominator'],
               axis='columns', inplace=True)
#test
#view the head of the data frame after removing the irrelevant colmuns
df1_clean.head()
```

## merge the last four columns into "stage' column
### replace all "None" values with empty string, then add the strings of the four columns. this should result in the stage value or an empty string if the stage is not given. If the same row has multipule stages, this should be removed later.

```python
# replace "None" with empty string
df1_clean.replace('None', value='', inplace=True)
# add the strings in the four columns and assign the results to the "srage" column
df1_clean['stage'] = df1_clean['doggo']+ df1_clean['floofer']+ df1_clean['pupper']+ df1_clean['puppo']
#drop the 4 columns
df1_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'],
              axis='columns', inplace=True)
#test
#view the haed of the data frame
df1_clean.head()
```

## Merging the data frames
### converting the data type of 'tweet_id' to string to be compatiable with the data in df_3
### the merge will be on the "tweet_id" column, olny the rows with common values is inculded.

```python
#merge df1_clean and df_2
merge=pd.merge(df1_clean,df_2, on='tweet_id')
#convert the data type of 'tweet_id' to string
merge['tweet_id'] = merge['tweet_id'].astype(str)
#merge the resultant with df_3
df = pd.merge(merge,df_3, on='tweet_id')
#test
#view the head of the merged dataframe
df.head()
```

```python
df.info()
```

# cleaning the merged df
### make a copy of the dataframe to make changes on.

```python
df_clean = df.copy()
#test
#view the head of the copied df
df_clean.head()
```

## convert 'timestamp' to datetime objecect
### use to_date method to convert the string in 'timestamp' column to datetime object

```python
df_clean['timestamp'] = pd.to_datetime(df_clean['timestamp'])
#test
#view the info of the datafram to check the datatype in 'timestamp'
df_clean.info()
```

## remove non-valied values of name
### replace the non-valied names with the string 'Not Given'

```python
df_clean['name'].replace(['a', 'an', 'the', ''], value='Not Given', inplace=True)
#test
#view the values in the 'name' colmun
df_clean['name'].value_counts()
```

## remove non-valied values of stage
### replace the non-valied stages (multipule stages)  with the string 'Not Given'

```python
df_clean['stage'].replace(['doggopupper','doggopuppo','doggofloofer' ,''], 'Not Given', inplace=True)
#test
#view the values in the 'stage' colmun
df_clean['stage'].value_counts()
```

## correcting wrong 'rating_numerator' data
### extract 'rating_numerator' form the 'text' column and replce the values in  'rating_numerator' with the extracted values

```python
#extract 'rating_numerator' value from the text on the row that have the rating in text
df = df_clean[df_clean.text.str.contains(r"(\d+\.\d*\/\d+)")][['text', 'rating_numerator']]
#locate the rows that need change in df_clean and assign values from the extrated df
df_clean.loc[df.index, 'rating_numerator'] = df['rating_numerator']
#test
#view the new values in 'rating_numerator' for the locations that have changes
df_clean.loc[df.index, 'rating_numerator']
```

## converting data type of 'rating_numerator
### Change the datatype from ineger to float

```python
df_clean['rating_numerator'] = df_clean['rating_numerator'].astype(float)
#test
#check the data type of 'rating_numerator' column in using info method
df_clean.info()
```

# sorting columns
### view the data in logical order

```python
#sorted list of columns
columns = ['tweet_id', 'timestamp', 'name', 'stage', 'rating_numerator', 'favourites_count',
          'retweet_count', 'img_num', 'p1', 'p1_conf', 'p1_dog', 'p2', 'p2_conf', 'p2_dog',
          'p3', 'p3_conf', 'p3_dog', 'source', 'text', 'expanded_urls','jpg_url']
#arrange the dataframe using the sorted list
df_clean = df_clean[columns]
#teat
#view the head of the dataframe
df_clean.head()
```
